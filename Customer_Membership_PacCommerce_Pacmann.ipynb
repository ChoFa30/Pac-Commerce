```python
from tabulate import tabulate
from math import sqrt
import math
```

```python
existing_users = {
    'Sumbul':'Platinum',
    'Ana':'Gold',
    'Cahya':'Platinum'
}
```

```python
class MembershipUser :
  data = existing_users
  member_tiers = {
      'Membership' :['Platinum', 'Gold', 'Silver'],
      'Discount' :[15, 10, 8],
      'Another Benefit' : ['Benefit Silver + Gold + Voucher Liburan + Cashback max. 30%', 'Benefit Silver + Voucher Ojek Online', 'Voucher Makanan']
  }
  tier_requirements = {
      'Membership' : ['Platinum','Gold','Silver'],
      'Monthly_Expense' : [8,6,5],
      'Monthly_Income' : [15,10,7]
  }

  def __init__(self, username, monthly_expense, monthly_income) :
    self.username = username
    self.monthly_expense = monthly_expense
    self.monthly_income = monthly_income

  def show_benefit(self) :
    print('Benefit :')
    print(tabulate(self.member_tiers, headers='keys', tablefmt='fancy_grid'))

  def show_requirements(self) :
    print('requirements :')
    print(tabulate(self.tier_requirements, headers='keys', tablefmt='fancy_grid'))

  def predict_membership(self, monthly_expense, monthly_income):
    expense = monthly_expense
    income = monthly_income
    try:
      platinum = round(sqrt(((expense-self.tier_requirements['Monthly_Expense'][0])**2) + ((income-self.tier_requirements['Monthly_Income'][0])**2)), 2)
      gold = round(sqrt(((expense-self.tier_requirements['Monthly_Expense'][1])**2) + ((income-self.tier_requirements['Monthly_Income'][1])**2)), 2)
      silver = round(sqrt(((expense-self.tier_requirements['Monthly_Expense'][2])**2) + ((income-self.tier_requirements['Monthly_Income'][2])**2)), 2)
      self.table = {
          'Membership' : ['Platinum', 'Gold', 'Silver'],
          'Eucledian Distance' : [platinum, gold, silver]
      }
      self.eucledian_distance = [platinum, gold, silver] # mencari nilai min
      self.membership_names = ['Platinum', 'Gold', 'Silver'] # bikin untuk predict eucledien
      min_eucledian = min(self.eucledian_distance) # mencari nilai min
      min_index = self.eucledian_distance.index(min_eucledian) # mencari index nilai min

      print('Hasil Perhitungan Eucledian Distance :')
      print(tabulate(self.table, headers='keys', tablefmt='fancy_grid'))
      print('')
      print(f'Eucledian Distance predict : {self.table['Membership'][min_index]}')
    except Exception as e:
      print(f'error : {e}')

  def calculate_price(self, membership, list_harga_barang) :
    self.membership = membership
    self.list_harga_barang = list_harga_barang
    membership_index = self.member_tiers['Membership'].index(membership.capitalize())
    get_data = self.member_tiers['Membership'][membership_index]
    print(get_data)
    sum_of_all = sum(list_harga_barang)
    print(sum_of_all)
    if get_data == 'Platinum' :
      diskon = 15/100*sum_of_all
      total = int(sum_of_all - diskon)
      print(f'Total harga yang harus dibayar : Rp{total:,}'.replace(',', '.'))
    elif get_data == 'Gold' :
      diskon = 10/100*sum_of_all
      total = int(sum_of_all - diskon)
      print(f'Total harga yang harus dibayar : Rp{total:,}'.replace(',', '.'))
    elif get_data == 'Silver' :
      diskon = 8/100*sum_of_all
      total = int(sum_of_all - diskon)
      print(f'Total harga yang harus dibayar : Rp{total:,}'.replace(',', '.'))
    else :
      print('diskon error')
```

```python
test1 = MembershipUser('Sumbul', 15, 20)

# attribute instance
print(f' Username : {test1.username}')
print(f' Monthly Expense : {test1.monthly_expense}')
print(f' Monthly Income : {test1.monthly_income}')
print('='*150)
print('')

# method show_benefit
print('Test Case 1 \n')
test1.show_benefit()
print('='*150)
print('')

# method requirements
print('Test Case 2 \n')
test1.show_requirements()
print('='*150)
print('')

# method predict_membership
print('Test Case 3 \n')
test1.predict_membership(2, 10)
print('='*150)
print('')

# method calculate_price
print('Test Case 4 \n')
test1.calculate_price('platinum', [150000, 100000, 100000])
print('='*150)
print('')
```
